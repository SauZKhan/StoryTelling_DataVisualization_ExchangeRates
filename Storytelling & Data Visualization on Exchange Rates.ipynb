# Storytelling and Data Visualization on Exchange Rates

#### The two branches of Data Visualization 
#### Exploratory data visualization (graph for ourselves to understand better & explore data) & 
#### Explainatory data viusalization (graph for others to understand, make a point or story telling)

#### In this project, we will work on explainatory data visualization

```python
import pandas as pd
import warnings

warnings.filterwarnings("ignore")
```

```python
exchange_rates = pd.read_csv("euro-daily-hist_1999_2022.csv")
exchange_rates.shape
```

```python
exchange_rates.head()
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar'}, inplace= True)
```

```python
exchange_rates.rename(columns={'Period\\Unit:': 'Time'}, inplace= True)
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
```

```python
exchange_rates.sort_values('Time', inplace=True)
```

```python
exchange_rates.head()
```

```python
exchange_rates.columns
```

```python
euro_to_dollar = exchange_rates[['Time', 'US_dollar']].copy()
euro_to_dollar['US_dollar'].value_counts() #62 '-' characters
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

## Rolling mean

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

#### If we look at the line's shape, we see many wiggles - rather than seeing a smooth line. The wiggles, however have meaning: they are the visual representation of the daily variation in the exchange rate. The rate goes up and downw, up and down again, day to day. The rate only shows clear upward or downward trends in the longer run(months or years).

#### Depending o nout goals we may not want to show that daily variation on our graph. If we want to hid it and show only the long term trends, we can use the rolling mean(also known as the moving average)

```python
values = pd.DataFrame()
values['daily_values'] = pd.Series(range(1,20,2))
values
```

```python
values['rolling_mean_2'] = values['daily_values'].rolling(2).mean()
values
```

```python
values['rolling_mean_3'] = values['daily_values'].rolling(3).mean()
values['rolling_mean_5'] = values['daily_values'].rolling(5).mean()

values
```

```python
plt.figure(figsize=(9,6))

plt.subplot(3,2,1)
plt.plot(euro_to_dollar['Time'],euro_to_dollar['US_dollar'] )
plt.title('Original values', weight='bold')

for i, rolling_mean in zip([2,3,4,5,6],
                          [7,30,50,100,365]):
    plt.subplot(3,2,i)
    plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar']. rolling(rolling_mean).mean())
    plt.title('Rolling Window:' + str(rolling_mean), weight='bold')
    
plt.tight_layout() #auto digits the padding between subplots
plt.show()
```

### Coming up with an idea

Here are the few story ideas for our data:
    - We show how the euro-dollar rate has changed during the covid pandemic. We can show the 2020 data and the 2016-2019 data as baseline. We can use a line plot.
    - We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for 2006 & 2009 for comparision. We can use a line plot.
    - We show comparitively how the euro-dollar rate changed under the last three US presidents (G. W Bush(2001-2009), Barack Obama(2009-2017) ad Donald Trump(2017-2021). We can use a line plot.
    

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['US_dollar'].rolling(30).mean()
euro_to_dollar
```

## Story telling Data Visualization

### Financial crisis 2007 - 2009

```python
financial_crisis = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2006) & (euro_to_dollar['Time'].dt.year <=2009)]
financial_crisis_7_8 = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2007) & (euro_to_dollar['Time'].dt.year <=2008)]
```

```python
import matplotlib.style as style
style.use('fivethirtyeight')

## Adding the plot
fig,ax = plt.subplots(figsize=(8,3))
ax.plot(financial_crisis['Time'], financial_crisis['rolling_mean'], linewidth=1, color='#A6D785')

## Highlighting the 2007-2008 period
ax.plot(financial_crisis_7_8['Time'], financial_crisis_7_8['rolling_mean'], linewidth=3, color='#e23d28')

ax.set_xticklabels([])

x = 0.02
for year in ['2006', '2007', '2008', '2009', '2010']:
    ax.text(x, -0.08, year, alpha=0.5, fontsize=11, transform= plt.gca().transAxes)
    x +=0.22888
    
# print (plt.yticks())

ax.set_yticklabels([])

y=0.07
for rate in ['1.2', '1.3', '1.4', '1.5']:
    ax.text(-.04, y, rate, alpha=0.5, fontsize=11, transform= plt.gca().transAxes)
    y +=0.2333
    
# Adding a title and a subtitle
ax.text(-.05, 1.2, "Euro-USD rate peaked at 1.59 during 2007-2008's financial crisis", weight ='bold', transform = plt.gca().transAxes)
ax.text(-.05, 1.1, "Euro-USD rate peaked at 1.59 during 2006-2010", size =12, transform = plt.gca().transAxes) 

### Adding a signature
ax.text(-.05, -0.25, 'TECHMA ZONE' + ' '*80 + 'Source: European Central Bank', color ='#f0f0f0', backgroundcolor = '#4d4d4d', size=12, transform = plt.gca().transAxes)

ax.axvspan(xmin=pd.to_datetime("2008-04-1"), xmax=pd.to_datetime("2008-09-1"), ymin=0.09, alpha=0.3, color='grey')

plt.show()
```
